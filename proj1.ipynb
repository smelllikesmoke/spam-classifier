```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')

#Ast solves this problem
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)   # Storing the list in clear format to movie genres col
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['crew'][0]
# list of dictionaries we only want to extract director title from the dictionaries
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())  #converting string into list so it can be appended
movies.head()
```

```python
# Now in cast we have full and last name splitted which could be interpreted as different keyword when theyre not
#so removing spaces
movies["genres"]=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies["keywords"]=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies["cast"]=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies["crew"]=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

Now we will check the similarity between 2 different tags for 2 different movies shown for recommendation

* We will use text vectorization technique "bag of words"
suppose in a 2d all lines represent each of the 5000 movies the closest vectors or lines will be selected for recommendation

\    |      /
 \   |    /
   \ |  /
     | _ _ _ _ _


* Bag of Words:
1- Combine all tags 
2- after concatenation df becomes large having many words 
3- Frequency of the desired word is calculated and extracted

for all tags (5000) each word frequency will be calculated and compared

* After above implementation, shape of data becomes:
(5000,5000)  # 5000 movies , 5000 words

* Stop words will be ignored as it holds no significance

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)

new_df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

* We have 4806 vectors spread across a 5000 dim space to get the closest vectors we will not use euclidean distance but calculate the closest angle

euclidean distance is not a good approach when dealing with higher dimensions thus, we calculate cosine distance
distance =k *1/similarity  inversely proportional
